```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sssd.core.layers.s4.hippo.normal_plus_low_rank import *
```

```python
def plot_transition_matrix(A, B):
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    
    # Plot heatmap of matrix A on the first subplot
    sns.heatmap(A, annot=False, cmap="YlGnBu", cbar=True, ax=ax[0])
    ax[0].set_title('Heatmap of Matrix A')
    
    # Plot the structure of B on the second subplot
    ax[1].plot(B, 'o-', label='B values')
    ax[1].set_title('Structure of B')
    ax[1].set_xlabel('Index')
    ax[1].set_ylabel('Value')
    ax[1].legend()
    ax[1].grid(True)
    
    # Show the combined plot
    plt.show()


def plot_low_matrix(P):
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Plot each row of the low-rank matrix P on separate subplots
    for i, row in enumerate(P):
        ax.plot(row, marker='o', label=f'Component {i}')
        
    ax.set_title('Line Plot of Low-Rank Matrix P')
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')
    ax.legend()
    ax.grid(True)
    plt.show()
```

### Scaled Legendre Measure

```python
A, B = LegsNormalPlusLowRank(matrix_size=32).generate_transition_matrix()
P = LegsNormalPlusLowRank(matrix_size=32, correction_rank=3).generate_low_rank_matrix()

plot_transition_matrix(A, B)
plot_low_matrix(P)
```

```python
A, B = LegtNormalPlusLowRank(matrix_size=32).generate_transition_matrix()
P = LegtNormalPlusLowRank(matrix_size=32, correction_rank=3).generate_low_rank_matrix()

plot_transition_matrix(A, B)
plot_low_matrix(P)
```
